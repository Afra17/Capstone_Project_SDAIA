```python
from markitdown import MarkItDown
import os

# 1. Initialize MarkItDown
md = MarkItDown()

# 2. Convert the PDF
# Make sure the path to your PDF is correct
result = md.convert(".\data\Mobile App Development.pdf")
raw_text = result.text_content

# 3. Define the Save Path
save_path = (".\data")

# Ensure the directory exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# 4. Save the output as a Markdown file
with open(save_path, "w", encoding="utf-8") as f:
    f.write(raw_text)

print(f"Done! File saved to: {save_path}")
```
